```python
# all imports
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode
from io import BytesIO
from pydub import AudioSegment
import os



# JavaScript to record audio in the browser
RECORD = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=30):
    display(Javascript(RECORD))
    print('Recording')
    s = output.eval_js('record(%d)' % (sec*1000))
    b = b64decode(s.split(',')[1])
    audio = AudioSegment.from_file(BytesIO(b))
    return audio

# Record for 30 seconds
audio = record()

# Split and save the audio into three 10-second segments
for i in range(3):
    start_time = i * 10000  # Start time in milliseconds
    end_time = (i + 1) * 10000  # End time in milliseconds
    segment = audio[start_time:end_time]
    segment.export(f"10-sec-{i+1}.wav", format="wav")

print("Audio segments saved successfully.")
```

```python
# all imports
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode
from io import BytesIO
from pydub import AudioSegment
import os

# JavaScript to record audio in the browser with a stop button
RECORD = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = () => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  window.recordStop = () => {
    recorder.stop()
  }
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
})
"""

def record(custom_folder_name, segment_length=10):
    # Display the recording interface
    display(Javascript(RECORD))
    print("Recording... Press the stop button in the output cell to stop recording.")
    output.eval_js('record()')

    # Wait for the user to stop the recording
    input("Press Enter in this cell to stop recording...")  # User presses Enter to stop
    output.eval_js('recordStop()')

    # Get the recorded audio
    s = output.eval_js('recordStop()')
    b = b64decode(s.split(',')[1])
    audio = AudioSegment.from_file(BytesIO(b))

    # Create the directory
    folder_path = f"/content/tortoise-tts/tortoise/voices/{custom_folder_name}"
    os.makedirs(folder_path, exist_ok=True)

    # Split and save the audio
    start_time = 0
    i = 0
    while start_time < len(audio):
        end_time = start_time + segment_length * 1000  # Convert to milliseconds
        segment = audio[start_time:end_time]
        segment.export(os.path.join(folder_path, f"{i}.wav"), format="wav")
        start_time = end_time
        i += 1

    print(f"Audio segments saved in {folder_path}")

# Example usage
record('Inesh2')
```

```python
from flask_ngrok import run_with_ngrok
from flask import Flask, request, jsonify, render_template_string
import os
from pydub import AudioSegment
from base64 import b64decode
from io import BytesIO

app = Flask(__name__)
run_with_ngrok(app)

@app.route("/")
def index():
    return """
    <h1>Audio Recorder</h1>
    <input type='text' id='voiceName' placeholder='Enter Voice Name'>
    <button onclick='startRecording()'>Start Recording</button>
    <button onclick='stopRecording()' style='display: none;' id='stopButton'>Stop Recording</button>
    <br><br>
    <a href='/read_pdfs'>Read PDFs</a>

    <script>
    var mediaRecorder;
    var audioChunks = [];

    function startRecording() {
        navigator.mediaDevices.getUserMedia({ audio: true })
            .then(stream => {
                mediaRecorder = new MediaRecorder(stream);
                mediaRecorder.ondataavailable = event => {
                    audioChunks.push(event.data);
                };
                mediaRecorder.start();
                document.getElementById('stopButton').style.display = 'block';
            });
    }

    function stopRecording() {
        mediaRecorder.stop();
        mediaRecorder.onstop = async () => {
            const audioBlob = new Blob(audioChunks);
            const audioUrl = URL.createObjectURL(audioBlob);
            const audioData = await fetch(audioUrl).then(r => r.blob()).then(blobToFile);

            // Send the audio data to Flask server
            const voiceName = document.getElementById('voiceName').value;
            fetch('/save_audio', {
                method: 'POST',
                headers: {
                    'Content-Type': 'application/json',
                },
                body: JSON.stringify({ voiceName: voiceName, audioData: audioData }),
            })
            .then(response => response.json())
            .then(data => {
                console.log(data.message);
            });

            audioChunks = [];
            document.getElementById('stopButton').style.display = 'none';
        };
    }

    function blobToFile(blob){
        return new Promise(resolve => {
            const reader = new FileReader();
            reader.onloadend = () => resolve(reader.result);
            reader.readAsDataURL(blob);
        });
    }
    </script>

    """

@app.route("/read_pdfs")
def read_pdfs():
    return """
    <h1>Read PDFs</h1>
    <form action="/upload_pdf" method="post" enctype="multipart/form-data">
        <input type="file" name="pdf_file">
        <input type="submit" value="Upload PDF">
    </form>
    <br>
    <button onclick="saveSelectedText()">Save Selected Text</button>
    <script>
    function saveSelectedText() {
        var selectedText = window.getSelection().toString();
        fetch('/save_selected_text', {
            method: 'POST',
            headers: {
                'Content-Type': 'application/json',
            },
            body: JSON.stringify({ text: selectedText }),
        })
        .then(response => response.json())
        .then(data => {
            console.log(data.message);
        });
    }
    </script>
    """

@app.route("/upload_pdf", methods=["POST"])
def upload_pdf():
    if 'pdf_file' not in request.files:
        return "No file part"
    file = request.files['pdf_file']
    if file.filename == '':
        return "No selected file"
    if file:
        filename = file.filename
        file.save(os.path.join("uploaded_pdfs", filename))
        return f"Uploaded {filename}"

@app.route("/save_selected_text", methods=["POST"])
def save_selected_text():
    text = request.json.get('text', '')
    with open("selectedtext.txt", "w") as file:
        file.write(text)
    return jsonify({"message": "Text saved successfully"})




@app.route("/save_audio", methods=["POST"])
def save_audio():
    data = request.json
    voice_name = data['voiceName']
    audio_data = data['audioData']

    # Decode the audio data
    audio_bytes = b64decode(audio_data.split(',')[1])
    audio = AudioSegment.from_file(BytesIO(audio_bytes))
    # Create directory for voice
    folder_path = f"/content/tortoise-tts/tortoise/voices/{voice_name}"
    os.makedirs(folder_path, exist_ok=True)

    # Split and save the audio
    segment_length = 10  # seconds
    start_time = 0
    i = 0
    while start_time < len(audio):
        end_time = start_time + segment_length * 1000  # Convert to milliseconds
        segment = audio[start_time:end_time]
        segment.export(os.path.join(folder_path, f"{i}.wav"), format="wav")
        start_time = end_time
        i += 1

    return jsonify({"message": "Audio saved successfully"})

@app.route("/get_details")
def get_details():
    return "<h1>Details page<h1>"

@app.route("/test")
def test_page():
    return "<h1>Test page<h1>"

if __name__ == "__main__":
    app.run()
```
