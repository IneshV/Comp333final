# Clean Files & Load HTML, CSS & JS

```python
import os
import shutil

def clean_folder(base_path, keep_folders):
    # List all items in the base directory
    for item in os.listdir(base_path):
        item_path = os.path.join(base_path, item)

        # Check if the item is a directory and not in the keep list
        if os.path.isdir(item_path) and item not in keep_folders:
            # Delete all files in the directory
            for file in os.listdir(item_path):
                file_path = os.path.join(item_path, file)
                if os.path.isfile(file_path):
                    os.remove(file_path)

            # Delete the directory
            shutil.rmtree(item_path)
            print(f"Deleted folder: {item_path}")

# Usage
base_path = '/content/tortoise-tts/tortoise/voices'
keep_folders = ['angie', 'emma', 'pat', 'tom', 'freeman']
clean_folder(base_path, keep_folders)
```

```python
from flask_ngrok import run_with_ngrok
from flask import Flask, request, jsonify, render_template_string, send_from_directory
import os
from pydub import AudioSegment
from base64 import b64decode
from io import BytesIO
from werkzeug.utils import secure_filename
from pdfminer.high_level import extract_text

def get_folders(path):
    return [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]

app = Flask(__name__)
run_with_ngrok(app)

UPLOAD_FOLDER = 'uploaded_pdfs'
os.makedirs(UPLOAD_FOLDER, exist_ok=True)
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

@app.route("/")
def index():
    folders = get_folders('/content/tortoise-tts/tortoise/voices')
    return render_template_string(INDEX_TEMPLATE, folders=folders)

@app.route("/read_pdfs", methods=["GET", "POST"])
def read_pdfs():
    folders = get_folders('/content/tortoise-tts/tortoise/voices')
    extracted_text = ""
    if request.method == "POST":
        pdf_file = request.files.get('pdf_file')
        if pdf_file and allowed_file(pdf_file.filename):
            filename = secure_filename(pdf_file.filename)
            file_path = os.path.join(app.config['UPLOAD_FOLDER'], filename)
            pdf_file.save(file_path)
            extracted_text = extract_text_from_pdf(file_path)
    return render_template_string(READ_PDFS_TEMPLATE, folders=folders, text=extracted_text)

def allowed_file(filename):
    return '.' in filename and filename.rsplit('.', 1)[1].lower() in {'pdf'}

def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

UPLOAD_FORM_TEMPLATE = """
<!doctype html>
<title>Upload PDF</title>
<h1>Upload PDF</h1>
<form method=post enctype=multipart/form-data>
  <input type=file name=pdf_file>
  <input type=submit value=Upload>
</form>
"""

TEXT_DISPLAY_TEMPLATE = """
<!doctype html>
<title>PDF Text</title>
<h1>Extracted Text from PDF</h1>
<div style="white-space: pre-wrap;">{{ text }}</div>
"""

@app.route("/save_audio", methods=["POST"])
def save_audio():
    data = request.json
    voice_name = data['voiceName']
    audio_data = data['audioData']

    # Decode the audio data
    audio_bytes = b64decode(audio_data.split(',')[1])
    audio = AudioSegment.from_file(BytesIO(audio_bytes))
    # Create directory for voice
    folder_path = f"/content/tortoise-tts/tortoise/voices/{voice_name}"
    os.makedirs(folder_path, exist_ok=True)

    # Split and save the audio
    segment_length = 10  # seconds
    start_time = 0
    i = 0
    while start_time < len(audio):
        end_time = start_time + segment_length * 1000  # Convert to milliseconds
        segment = audio[start_time:end_time]
        segment.export(os.path.join(folder_path, f"{i}.wav"), format="wav")
        start_time = end_time
        i += 1

    # Inside the save_audio() function, after saving the audio successfully:
    return jsonify({"message": "Audio saved successfully", "folders": get_folders('/content/tortoise-tts/tortoise/voices')})

@app.route("/read_text", methods=["POST"])
def read_text():
    data = request.json
    text = data['text']
    person = data['person']
    speed = data['speed']

    try:
        gen_speech(text, person, speed)
        return jsonify({"message": "Speech generated successfully"})
    except Exception as e:
        return jsonify({"error": str(e)})

@app.route('/get_audio', methods=['POST'])
def get_audio():
    filename = request.json.get('filename')
    try:
        # Ensure the directory path is correct
        directory_path = '/content/tortoise-tts'
        return send_from_directory(directory=directory_path, path=filename, as_attachment=True)
    except FileNotFoundError:
        return jsonify({"error": "File not found"}), 404



def gen_speech(text, person, speed):
    voice_samples, conditioning_latents = load_voice(person)

    gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                              preset=speed)
    torchaudio.save(f'generated-{person}.wav', gen.squeeze(0).cpu(), 24000)
    IPython.display.Audio(f'generated-{person}.wav')


if __name__ == "__main__":
    app.run()
```
